# CFDE-IDG Pharos DiseasePage STRIDES Demo

* https://www.nih-cfde.org/
* https://druggablegenome.net

In this simple CFDE-IDG demo we pull Pharos/TCRD data from AWS S3 cloud storage provided via NIH STRIDES, for display, analysis and integration with related entities and datasets.

## Imports

```
import sys,os,io,json,requests,urllib.request
import pandas as pd
```

## Configuration

Specify disease of interest, S3 URL. CFDE Portal links: [Disease](https://app.nih-cfde.org/chaise/record/#1/CFDE:disease/nid=16) | [IDG Collection](https://app.nih-cfde.org/chaise/record/#1/CFDE:collection/nid=28014) | [IDG File](https://app.nih-cfde.org/chaise/record/#1/CFDE:file/nid=634970)


```
bucket_id = "kmc-idg"
bucket_host = f"{bucket_id}.s3.amazonaws.com"
url_basepath = "unm/tcrd"
doid = 2841
```

## Request DiseasePage file from AWS S3 bucket

```
filename = f"tcrd_disease_DOID_{doid:06d}.json"
response = requests.get(f"https://{bucket_host}/{url_basepath}/{filename}")
diseasepage = response.json()
pd.DataFrame.from_dict({key:[str(val)] for key,val in diseasepage.items()}).transpose()
```

## Associations

```
associations = pd.DataFrame.from_dict(diseasepage['associations'])
associations
```
